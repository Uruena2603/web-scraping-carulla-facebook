```python
#DIA 1
from bs4 import BeautifulSoup
import pandas as pd

# Ruta del archivo HTML
ruta_archivo = "D:/User/Desktop/carulla1.html"

# Leer el archivo HTML 
with open(ruta_archivo, "r", encoding="utf-8") as file:
    html_content = file.read()

# Crear el objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Encontrar los nombres de los productos
nombres = [span.text.strip() for span in soup.find_all('span', class_='vtex-store-components-3-x-productBrand')]

# Encontrar los precios de los productos
precios = []
for div in soup.find_all('div', class_='exito-vtex-components-4-x-PricePDP'):
    precio_span = div.find('span', class_='exito-vtex-components-4-x-currencyContainer')
    if precio_span:
        precio = precio_span.text.strip()
        precios.append(precio)
    else:
        precios.append(None)

# Crear un dataframe con los datos
tabla = {'Producto': nombres, 'Precio': precios}
dia1 = pd.DataFrame(tabla)

# Mostrar el dataframe
print("5 de mayo de 2024")
print(dia1)
```

```python
#DIA 2

# Ruta del archivo HTML
ruta_archivo2 = "D:/User/Desktop/carulla2.html"

# Leer el archivo HTML 
with open(ruta_archivo2, "r", encoding="utf-8") as file:
    html_content = file.read()

# Crear el objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Encontrar los nombres de los productos
nombres2 = [span.text.strip() for span in soup.find_all('span', class_='vtex-store-components-3-x-productBrand')][:21]

# Encontrar los precios de los productos
precios2 = []
for div in soup.find_all('div', class_='exito-vtex-components-4-x-PricePDP')[:21]:
    precio_span2 = div.find('span', class_='exito-vtex-components-4-x-currencyContainer')
    if precio_span2:
        precio2 = precio_span2.text.strip()
        precios2.append(precio2)
    else:
        precios2.append(None)

# Crear un dataframe con los datos
tabla2 = {'Producto': nombres2, 'Precio': precios2}
dia2 = pd.DataFrame(tabla2)

# Mostrar el dataframe
print("6 de mayo de 2024")
print(dia2)
```

```python
#DIA 3

# Ruta del archivo HTML
ruta_archivo3 = "D:/User/Desktop/carulla3.html"

# Leer el archivo HTML 
with open(ruta_archivo3, "r", encoding="utf-8") as file:
    html_content = file.read()

# Crear el objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Encontrar los nombres de los productos
nombres3 = [span.text.strip() for span in soup.find_all('span', class_='vtex-store-components-3-x-productBrand')][:21]

# Encontrar los precios de los productos
precios3 = []
for div in soup.find_all('div', class_='exito-vtex-components-4-x-PricePDP')[:21]:
    precio_span3 = div.find('span', class_='exito-vtex-components-4-x-currencyContainer')
    if precio_span3:
        precio3 = precio_span3.text.strip()
        precios3.append(precio3)
    else:
        precios3.append(None)

# Crear un dataframe con los datos
tabla3 = {'Producto': nombres3, 'Precio': precios3}
dia3 = pd.DataFrame(tabla3)

# Mostrar el dataframe
print("7 de mayo de 2024")
print(dia3)
```

```python
# Combinar los DataFrames dia1, dia2 y dia3
union = pd.concat([dia1, dia2, dia3])

# Agrupar por Producto y Mostrar Precios en los 3 días
tablaFinal = union.groupby('Producto').agg({'Precio': lambda x: list(x)})

# Imprimir la tabla final con separadores entre filas
print("Precios de Productos en 3 Días")
print("-" * 50)
for producto, precios in tablaFinal.iterrows():
    print(f"Producto: {producto}")
    print(f"Precios: {', '.join([str(p) for p in precios['Precio']])}")
    print("-" * 50)
```
