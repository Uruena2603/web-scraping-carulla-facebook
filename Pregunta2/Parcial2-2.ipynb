```python
import pandas as pd

# Ruta de la carpeta que contiene los archivos
carpeta = 'D:/User/Desktop/UCP/SEMESTRE 7/Python/Parcial2/Facebook/'

# Inicializar un diccionario para combinar los datos
diccionario_combinado = {}
j = 0

# Iterar sobre los archivos en la carpeta
for i in range(1, 11):  # Asumiendo que tengo 10 archivos
    nombre_archivo = os.path.join(carpeta, f'archivo{i}.txt')

    # Leer el archivo
    with open(nombre_archivo, 'r', encoding='utf-8') as archivo:
        contenido = archivo.read()

    # Convertir el contenido a diccionario
    splitado = contenido.split("\n")
    for data in splitado:
        data = data.replace("null", "None")
        data = data.replace('\n', '')
        data = data.replace('datetime.datetime', 'datetime')
        try:
            diccionario_combinado[j] = eval(data)
            j += 1
        except Exception as e:
            pass

# Crear un DataFrame a partir del diccionario combinado
df = pd.DataFrame.from_dict(diccionario_combinado).transpose()

# Mostrar las primeras y √∫ltimas filas del DataFrame para verificar
#df.head()
df
```

```python
# Contar comentarios por seguidor
comentarios_por_seguidor = df['commenter_name'].value_counts().reset_index()
comentarios_por_seguidor.columns = ['Seguidor', 'N√∫mero de Comentarios']

# Seleccionar los 10 seguidores que m√°s comentaron
seguidores_comentarios_mas = comentarios_por_seguidor.head(10)

# Mostrar los seguidores que comentaron en m√°s publicaciones
print('a. Seguidores que comentaron en m√°s publicaciones')
for index, row in seguidores_comentarios_mas.iterrows():
    print(f"- {row['Seguidor']}: {row['N√∫mero de Comentarios']} comentarios")

print() 
```

```python
# Convertir las fechas de los comentarios a datetime
df['comment_time'] = pd.to_datetime(df['comment_time'])

# Extraer la hora de cada comentario
df['hora_comentario'] = df['comment_time'].dt.hour

# Calcular la hora promedio
hora_promedio = df['hora_comentario'].mean()
print("La hora promedio donde los seguidores comentaron m√°s es:", hora_promedio)

print()
```
